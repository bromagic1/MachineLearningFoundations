
## Computing Gradients


```python

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
```

## Function

```python

def f(w1, w2):
    return 3 * w1 ** 2 + 2 * w1 * w2
```

```python



x1 = np.arange(0, 10)
x2 = np.arange(0, 10)
print(x1)
print(x2)

```

```python
 f(5, 3)
```

## Plotting

```python


B, D = np.meshgrid(x1, x2)
nu = f(B, D)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(B, D, nu)
plt.xlabel('w1')
plt.ylabel('w2')
plt.show()
```

```python
plt.contourf(B, D, nu)
plt.colorbar()
plt.xlabel('w1')
plt.ylabel('w2')
plt.show()
```

## Calculate Gradients

```python
w1 = tf.Variable(5.)
```

```python
w2 = tf.Variable(3.)
```

```python

with tf.GradientTape() as tape:
    z = f(w1, w2)
    
gradients = tape.gradient(z, [w1, w2])
```
